```python
df = pd.read_csv('./station.csv')
df.head()

dw = pd.read_csv('./weather.csv')
dw.head()


dtp = pd.read_csv('./trip.csv', error_bad_lines=False)

dtp = dtp.drop_duplicates()
dtp.head()


# The error below is not my fault

```

```python
# 1. What is the average trip duration for a borrowed bicycle?

pd.DataFrame(dtp['tripduration']).mean()
```

```python
# 2. What's the most common age of a bicycle-sharer?

pd.DataFrame(dtp['birthyear']).groupby(dtp.birthyear).count().sort_values('birthyear')
# most common birthyear is 1987
# 2018 - 1987
# answer: 31 - Millenials
```

```python
# 3. Given all the weather data here, find the average precipitation per month, and the median precipitation.
dw.head()

# avg precip
dw['Date'] = pd.to_datetime(dw['Date'], errors ='coerce')
dw['MONTH'] = dw['Date'].dt.month
dw['Precipitation_In'].groupby(dw.MONTH).mean()
```

```python
# med precip
dw['Precipitation_In'].groupby(dw.MONTH).median()
```

```python
# 4. What’s the average number of bikes at a given bike station?

pd.DataFrame(df['current_dockcount']).mean()
# answer = 16.5
```

```python
# 5. When a bike station is modified, is it more likely that it’ll lose bikes or gain bikes? How do you know?

pd.DataFrame(df[['install_dockcount']]).mean()
# 17.58
pd.DataFrame(df[['current_dockcount']]).mean()
# 16.51
# answer = it is more likely to lose bikes
```

```python
# 6. Come up with 3 more questions that can be answered with this data set.

# What are the top 5 installed bicycle dock counts?
pd.DataFrame(df[['install_dockcount']]).sort_values('install_dockcount', ascending = False).head()
```

```python
# What are the most common weather conditions?
pd.DataFrame(dw['Events']).groupby(dw.Events).count().sort_values('Events', ascending=False).head()
```
